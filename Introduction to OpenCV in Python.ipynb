```python
ret,thresh1 = cv2.threshold(RGB_img,127,255,cv2.THRESH_BINARY)
ret,thresh2 = cv2.threshold(RGB_img,127,255,cv2.THRESH_BINARY_INV)
ret,thresh3 = cv2.threshold(RGB_img,127,255,cv2.THRESH_TRUNC)
ret,thresh4 = cv2.threshold(RGB_img,127,255,cv2.THRESH_TOZERO)
ret,thresh5 = cv2.threshold(RGB_img,127,255,cv2.THRESH_TOZERO_INV)

titles = ['Original Image','BINARY','BINARY_INV','TRUNC','TOZERO','TOZERO_INV']
images = [RGB_img, thresh1, thresh2, thresh3, thresh4, thresh5]

for i in range(6):
    plt.subplot(2,3,i+1),plt.imshow(images[i],'gray')
    plt.title(titles[i])
    plt.xticks([]),plt.yticks([])

plt.show()
```

```python
dst = cv2.fastNlMeansDenoisingColored(RGB_img,None,10,10,7,21)

plt.subplot(121),plt.imshow(RGB_img)
plt.subplot(122),plt.imshow(dst)
plt.show()
```

```python
rows,cols = RGB_img.shape[:2] 
M = cv2.getRotationMatrix2D((cols/2,rows/2),90,1) 
dst = cv2.warpAffine(RGB_img,M,(cols,rows)) 
plt.imshow(dst)
```

```python
M = np.float32([[1,0,-100],[0,1,-100]]) 
dst = cv2.warpAffine(RGB_img,M,(cols,rows)) 
plt.imshow(dst)
```
