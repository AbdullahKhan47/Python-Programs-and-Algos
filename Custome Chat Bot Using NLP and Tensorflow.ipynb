```python
words=[]
word_doc=[]
word_doc2=[]
labels=[]
```

```python
for intent in data['intents']:
    if intent['tag'] not in labels:
        labels.append(intent['tag'])
    for pattern in intent['patterns']:
        print(pattern)
        w = nltk.word_tokenize(pattern)
        words.extend(w)
        word_doc.append(w)
        word_doc2.append(intent["tag"])
```

```python
training = []
output= []
for x,doc in enumerate(word_doc):
    bag=[]
    all_words = [stemmer.stem(w.lower()) for w in doc]
    for w in words:
        if w in all_words:
            bag.append(1)
        else:
            bag.append(0)
    print(doc)
    print(bag)
    output_row = out_epmty[:]
    output_row[labels.index(word_doc2[x])] = 1
    print(output_row)
    training.append(bag)
    output.append(output_row)
```

```python
def serve(u_input):
    bag = [0 for _ in range(len(words))]
    input_words = nltk.word_tokenize(u_input)
    input_words = [stemmer.stem(word.lower()) for word in input_words]
    for u_words in input_words:
        for idx,w in enumerate(words):
            if w== u_words:
                bag[idx] = 1
    return np.array(bag)
```

```python
def chat():
    print("Welcome to the Foodpanda")
    while True:
        user_input = input("you: ")
        if user_input.lower()=="quit":
            break
        results = model.predict([serve(user_input)])
        results_index = np.argmax(results)
        tag = labels[results_index]
        for tg in data["intents"]:
            if tg['tag'] == tag:
                responses = tg['responses']
        print(random.choice(responses))
```
